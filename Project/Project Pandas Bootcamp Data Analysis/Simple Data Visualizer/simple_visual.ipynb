## Visualisation Dataset

### Line Chart

Penjelasan kode:

sns.lineplot(...)

Here are the parameters used:
- data=df: The data used comes from DataFrame df.
- x=x_line, y=y_line: The x and y axes are taken from the x_line and y_line columns in the DataFrame.
- hue=‘subject’: The colour of the line will be differentiated based on the category in the subject column.
- style=‘subject’: The style of the line (e.g. solid, dashed) will also be distinguished based on the category in the subject column.
- palette=‘viridis’: The colour scheme used is viridis.
- sort=True: The data will be sorted by the value on the x-axis before drawing.


```python
# Line chart

x_line = 'level'
y_line = 'price'

plt.figure(figsize=(8,4)) # lebar 8, tinggi 4
sns.lineplot(data=df, x=x_line, y=y_line, hue='subject', style='subject', palette='viridis', sort=True)
plt.title(f'Line chart of {x_line}')
plt.xlabel(x_line)
plt.ylabel(y_line)
plt.show()

# Save result visualisation
plt.savefig('line_chart_level.png', dpi=300, bbox_inches='tight')
plt.show()
```

### Bar Graph

sns.barplot(data=df, x=x_bar, y=y_bar, palette='viridis'):

Parameter:
- data=df: The DataFrame that contains the data.

- x=x_bar: The column that will be used as the x-axis (subject).

- y=y_bar: The column that will be used as the y-axis (content_duration).

- palette=‘viridis’: The colour scheme used for the bar plot. viridis is one of Seaborn's built-in palettes that has good colour gradation.

```python
# Bar graph
x_bar = 'subject'
y_bar = 'content_duration'

plt.figure(figsize=(8,4)) # lebar 8, tinggi 4
sns.barplot(data=df, x=x_bar, y=y_bar, palette='viridis') # 
plt.title(f'Bar Plot {x_bar}')
plt.xlabel(x_bar)
plt.ylabel(y_bar)
plt.show()

# Save Visualisation Result
plt.savefig('subject_course.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Visualisasi dengan Matplotlib

x_data = 'subject'
y_data = 'content_duration'

fig, ax = plt.subplots(figsize=(8,4))
colors = sns.color_palette('viridis') # 
ax.bar(df[x_data], df[y_data], color=colors)
ax.set_ylabel(y_data)
ax.set_title(f'Bar Chart {x_data}')
ax.legend(title=x_data)
plt.show()
```

```python
plt.figure(figsize=(8,6))
df['is_paid'].value_counts().plot(kind='bar') # retrieves the is_paid column data, and uses value counts for visual data counts
plt.title(f'Paid vs Unpaid Course')
plt.xlabel('Is_Paid', fontsize=12)
plt.ylabel('Course Count')
plt.show()

# Save visualisation results
plt.savefig('paid_unpaid_course.png', dpi=300, bbox_inches='tight')
plt.show()
```

Information:
- Almost all courses are paid and only a few courses are free.

### Histogram Graph

sns.histplot(data=df, x=x, bins=10, kde=True, color='red', edgecolor='black')
Parameter:
- data=df: The data used is DataFrame df.

- x=x: The column to be plotted is the ‘level’ column (the value of the x variable).

- bins=10: The histogram will be divided into 10 bins (interval groups).

- kde=True: Adds a Kernel Density Estimate (KDE) plot on top of the histogram to estimate the distribution of the data.

- colour=‘red’: The histogram colour is set to red.

- edgecolour=‘black’: The border of each bin is set to black.

plt.savefig(‘Histogram_graph_subject.png’, dpi=300, bbox_inches=‘tight’)

Parameter:
- Saves the plot into an image file with the name ‘Histogram_graph_subject.png’.
- dpi=300: Sets the image resolution to 300 dots per inch (DPI). The higher the DPI, the higher the image quality.
- bbox_inches=‘tight’: Ensures that the entire plot (including titles and labels) fits into the image without being cut off.

```python
x = 'level'

plt.figure(figsize=(8, 4))
sns.histplot(data=df, x=x, bins=10, kde=True, color='red', edgecolor='black')
plt.title(f'Histogram Graph {x}')
plt.xlabel(x)
plt.ylabel('Frequency')
plt.show

# Memnyimpan hasil visualisasi
plt.savefig('hist_levelCourse.png', dpi=300, bbox_inches='tight')
plt.show()
```

Information:
- The course levels that users are interested in are All Levels and Beginner Level.

### Histogram Graph with matplotlib
- plt.hist(): Creates a histogram.

- bins=30: Divides the data into 30 intervals.

- colour=‘blue’: The colour of the histogram.

- edgecolour=‘black’: The colour of the edges of the histogram bars.

- alpha=0.7: The transparency level of the bars.

- plt.xlabel() and plt.ylabel(): Adds x and y axis labels.

- plt.title(): Adds the plot title.

```python
hist_plt = 'subject'

plt.figure(figsize=(8,6))
plt.hist(df[hist_plt], bins=10, edgecolor='black', alpha=0.7)
plt.title(f'Histogram Graph {hist_plt}')
plt.xlabel(hist_plt)
plt.ylabel('Frequency')

# Save result viusalisation
plt.savefig('Histogram_graph_subject.png', dpi=300, bbox_inches='tight')
plt.show()
```

Information:
- The data shows that the subjects business finance and web development are popular with students.
- Graphic design has the least number of students followed by musical instruments.
